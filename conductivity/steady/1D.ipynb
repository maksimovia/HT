```python
#1D стационар ГУ-1
import numpy as np
import matplotlib.pyplot as plt

T1 = 50
T2 = 70
L = 1
M = 5
lamda = 50
qv = 10000

dx = L/M

A = np.zeros((M,M))
B = np.zeros(M)

for i in range(1,M-1):
    A[i,i] = -2
    A[i,i-1] = 1
    A[i,i+1] = 1
    B[i] = -qv*dx**2/lamda
    
A[0,0] = 1
B[0] = T1

A[-1,-1] = 1
B[-1] = T2

T = np.linalg.solve(A,B)
print(T)

X = [dx/2+i*dx for i in range(M)]
plt.plot(X,T)
plt.show()
```

```python
#1D стационар ГУ-3
import numpy as np
import matplotlib.pyplot as plt

Tf1 = 50
alpha1 = 15

Tf2 = 70
alpha2 = 10

L = 1
M = 5
lamda = 50
qv = 10000

dx = L/M

A = np.zeros((M,M))
B = np.zeros(M)

for i in range(1,M-1):
    A[i,i] = -2
    A[i,i-1] = 1
    A[i,i+1] = 1
    B[i] = -qv*dx**2/lamda
    
A[0,0] = -(1+alpha1*dx/lamda)
A[0,1] = 1
B[0] = -alpha1*dx/lamda*Tf1 - qv*dx**2/lamda

A[-1,-1] = -(1+alpha2*dx/lamda)
A[-1,-2] = 1
B[-1] = -alpha2*dx/lamda*Tf2 - qv*dx**2/lamda

T = np.linalg.solve(A,B)
print(T)

X = [dx/2+i*dx for i in range(M)]
plt.plot(X,T)
plt.show()
```
