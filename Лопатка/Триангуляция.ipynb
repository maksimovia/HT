```python
import matplotlib.pyplot as plt
import pandas as pd

blade_points = pd.read_excel("blade.xlsx", sheet_name = "blade")
blade_lines  = pd.DataFrame(columns=['X1', 'Y1', 'X2','Y2','L'])

for i in range(len(blade_points)-1):
    X1 = blade_points.loc[i,'X']
    Y1 = blade_points.loc[i,'Y']
    X2 = blade_points.loc[i+1,'X']
    Y2 = blade_points.loc[i+1,'Y']
    L = ((X2-X1)**2 + (Y2 - Y1)**2)**0.5
    blade_lines.loc[i] = [X1,Y1,X2,Y2,L]

plt.figure(figsize=(16, 16))
for i in range(len(blade_lines)):
    plt.plot([blade_lines.loc[i,'X1'],blade_lines.loc[i,'X2']],[blade_lines.loc[i,'Y1'],blade_lines.loc[i,'Y2']], color = 'black')
plt.xlim([-0.25,1.25])
plt.ylim([-0.75,0.75])
plt.show()





blade_lines
```
