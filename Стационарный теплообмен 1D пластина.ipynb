```python
#1D стационар ГУ-1
import numpy as np
import matplotlib.pyplot as plt

T1 = 50
T2 = 70
L = 1
M = 5
lamda = 50
qv = 10000

dx = L/(M-1)

A = np.zeros((M,M))
B = np.zeros(M)

for i in range(1,M-1):
    A[i,i] = -2
    A[i,i-1] = 1
    A[i,i+1] = 1
    B[i] = -qv*dx**2/lamda
    
A[0,0] = 1
B[0] = T1

A[M-1,M-1] = 1
B[M-1] = T2

T = np.linalg.solve(A,B)
print(T)

X = [i*dx for i in range(M)]
plt.plot(X,T)
plt.show()
```

```python
#1D стационар ГУ-2
import numpy as np
import matplotlib.pyplot as plt

T1 = 50
q2 = 1000

L = 1
M = 5
lamda = 50
qv = 10000

dx = L/(M-1)

A = np.zeros((M,M))
B = np.zeros(M)

for i in range(1,M-1):
    A[i,i] = -2
    A[i,i-1] = 1
    A[i,i+1] = 1
    B[i] = -qv*dx**2/lamda
    
A[0,0] = 1
B[0] = T1

A[M-1,M-1] = -1
A[M-1,M-2] = 1
B[M-1] = -q2*dx/lamda - qv*dx**2/(2*lamda)

T = np.linalg.solve(A,B)
print(T)

X = [i*dx for i in range(M)]
plt.plot(X,T)
plt.show()
```

```python
#1D стационар ГУ-3
import numpy as np
import matplotlib.pyplot as plt

# T1 = 0
Tf1 = 50
alpha1 = 15

Tf2 = 70
alpha2 = 10

L = 1
M = 5
lamda = 50
qv = 10000

dx = L/(M-1)

A = np.zeros((M,M))
B = np.zeros(M)

for i in range(1,M-1):
    A[i,i] = -2
    A[i,i-1] = 1
    A[i,i+1] = 1
    B[i] = -qv*dx**2/lamda
    
# A[0,0] = 1
# B[0] = T1
A[0,0] = -(1+alpha1*dx/lamda)
A[0,1] = 1
B[0] = -alpha1*dx*Tf1/lamda - qv*dx**2/(2*lamda)

A[M-1,M-1] = -(1+alpha2*dx/lamda)
A[M-1,M-2] = 1
B[M-1] = -alpha2*dx*Tf2/lamda - qv*dx**2/(2*lamda)

T = np.linalg.solve(A,B)
print(T)

X = [i*dx for i in range(M)]
plt.plot(X,T)
plt.show()
```
